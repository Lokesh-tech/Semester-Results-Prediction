```
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from numpy import array
from matplotlib import pyplot
```

```
dataset = loadtxt('semester.csv',delimiter=',')
```

```
dataset
```

```
X = dataset[:60,0:4]
Y = dataset[:60,4]
```

```
X
```

```
Y
```

```
model = Sequential()
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))
...
```

```
model.compile(loss='mse', optimizer='adam',metrics=['msle'])
```

```
history=model.fit(X, Y, epochs=1000, verbose=2)
```

```
X_new=dataset[60:,0:4]
```

```
y_new=model.predict(X_new)
```

```
print("X=%s, Predicted=%s" % (X_new, y_new))
```

```
pyplot.plot(history.history['msle'])
pyplot.show()
```
